# Interpolation
## Is a method to connect input data points with a goal to best approximate missing values. 

##    Linear Interpolation: connects points by straight line, 
*    Any two points in space $(x_1, y_1), (x_2, y_2)$ can be connected by a straight line using point slope equation.
*    Slope of this line is given by $\dfrac{(y_2-y_1)}{(x_2-x_1)}$ 
*    Equation of the line is $y(x) = y_1 + \dfrac{(y_2-y_1)}{(x_2-x_1)} * (x-x_1)$

```python
import numpy as np
import matplotlib.pyplot as plt
# matplotlib tutorials:-  https://matplotlib.org/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py
from scipy import interpolate
# %matplotlib qt 
# %matplotlib inline 
# %matplotlib notebook

x = np.arange(10)
y = np.sin(x)
plt.plot(x, y, 'o')
plt.title("Data Points")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

```python
x = np.linspace(0,9,10)
y = np.sin(x)
plt.plot(x, y, 'o', label='points')

linear = interpolate.interp1d(x, y)
xmissing = np.linspace(0, 9, 50)
ylinear = linear(xmissing)

plt.plot(xmissing, ylinear, '*', label='linear')
plt.plot(xmissing, np.sin(xmissing), '-', label='sin')
plt.legend(loc='lower left', ncol=2)
plt.title("Linear Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

### Another Example if the input points are known to have exponential relationship

```python
x = np.arange(10)
yexp = np.exp(-x/2.0)
plt.plot(x, yexp, 'o', label = 'points')
plt.legend(loc='lower left', ncol=2)
plt.title("Data Points")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

```python
# lets try to fit linear on these points.



x = np.arange(10)
yexp = np.exp(-x/2.0)
plt.plot(x, yexp, 'o', label = 'points')

linear = interpolate.interp1d(x, yexp)
xmissing = np.arange(0, 9, 0.1)
yexplinear = linear(xmissing)

plt.plot(xmissing, yexplinear, '*', label='linear')
plt.plot(xmissing, np.exp(-xmissing/2.0), '--', label='exp')
plt.legend(loc='lower left', ncol=2)
plt.title("Linear Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```

```python
# lets try to fit log linear on these points.
#%matplotlib qt 
#%matplotlib inline 
x = np.arange(10)
yexp = np.exp(-x/2.0)
plt.plot(x, yexp, 'o', label = 'points')

linear = interpolate.interp1d(x, yexp)
xmissing = np.arange(0, 9, 0.1)
yexplinear = linear(xmissing)
plt.plot(xmissing, yexplinear, '*', label='linear')

logyexp = np.log(yexp)
loglinear = interpolate.interp1d(x, logyexp)
yExpLogLinear = loglinear(xmissing)
plt.plot(xmissing, np.exp(yExpLogLinear), '*', label='Log Linear')

plt.plot(xmissing, np.exp(-xmissing/2.0), '--', label='exp')
plt.legend(loc='lower left', ncol=2)
plt.title("LogLinear Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values');
#plt.xlim(-0.5, 1.5)
```

```python
from scipy.interpolate import CubicSpline

x = np.arange(10)
y = np.sin(x)
plt.plot(x, y, 'o', label='points')

linear = interpolate.interp1d(x, y)
xmissing = np.arange(0, 9, 0.1)
ylinear = linear(xmissing)
plt.plot(xmissing, ylinear, '*', label='linear')

cSpline = CubicSpline(x, y)
ySpline = cSpline(xmissing)
plt.plot(xmissing, ySpline, '*', label='Cubic Spline')

#plt.plot(xmissing, np.sin(xmissing), '-', label='sin')
plt.legend(loc='lower left', ncol=2)
plt.title("Cubic Spline Interpolation")
plt.xlabel('X-values')
plt.ylabel('Y-values');
```
